```python
import pandas as pd

incomes = pd.read_csv("../data/raw/LAPI 1124.csv")
incomes.head()
incomes.iloc[110:120]
```

```python
state_names=["Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado", "Connecticut", "Delaware", "Florida", "Georgia", "Hawaii", "Idaho", "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky", "Louisiana", "Maine", "Maryland", "Massachusetts", "Michigan", "Minnesota", "Mississippi", "Missouri", "Montana", "Nebraska", "Nevada", "New Hampshire", "New Jersey", "New Mexico", "New York", "North Carolina", "North Dakota", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Vermont", "Virginia", "Washington", "West Virginia", "Wisconsin", "Wyoming"]

incomes['state'] = None
current_state = None
for idx, row in incomes.iterrows():
    name = str(row[0]).strip()
    rank = str(row[4]).strip()
    if name in state_names and rank == "--":
        current_state = name
        incomes.at[idx, 'state'] = current_state
    else:
        incomes.at[idx, 'state'] = current_state

# Drop rows where the County column is a state name
incomes = incomes[
    (incomes['Rank in state 2023'] != '--')
]
incomes.head()
```

```python
incomes.iloc[:120]
```

```python
incomes.columns = ['County', 'Drop1', 'Drop2', 'Income', 'Rank', 'Drop4', 'Drop5', 'Drop6', 'State']

incomes.drop(columns = ['Drop1', 'Drop2', 'Drop4', 'Drop5', 'Drop6', 'Rank'], inplace = True)

#clear all blank rows
incomes = incomes.dropna(subset=['Income'])
print(incomes.isna().sum())
incomes['Income'] = incomes['Income'].str.replace(',', '').astype(int)

incomes.head()
```

```python
#Add fips
from addfips import AddFIPS
af = AddFIPS()

incomes["FIPS"] = incomes.apply(
    lambda row: af.get_county_fips(state=row["State"], county=row["County"]),
    axis=1
)
```

```python
incomes.to_csv("../data/cleaned/incomes_cleaned.csv")
```
