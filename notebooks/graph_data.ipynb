```python
import seaborn as sns
import pandas as pd
```

```python
sns.set_palette('dark')
df = pd.read_csv("income-aqi.csv")
df = df[df["Income"] <= 200000]
ax = sns.regplot(x = df['Income'], y = df['Median AQI'])
ax.set_title("Linear Regression Plot of County Income vs County Median AQI in the US in 2023")
print(df['Income'].corr(df['Median AQI']))
```
